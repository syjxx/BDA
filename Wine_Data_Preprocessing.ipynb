```python
df.dropna(inplace=True)
```

```python
df = df[['alcohol', 'ash', 'nonflavanoid_phenols' ,'proanthocyanins']]
```

#스케일링 진행

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler
```

###MinMaxScaler

```python
mm_sca = MinMaxScaler()
mm_sca.fit(df)
```

```python
tr_mm_sca = mm_sca.transform(df)
tr_mm_sca
```

```python
df_mm_sc= pd.DataFrame(tr_mm_sca , columns = df.columns)
df_mm_sc
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,init='k-means++', max_iter=300,random_state=111)
kmeans.fit(df_mm_sc)
print(kmeans.labels_)
```

```python
df_mm_sc['cluster'] = kmeans.labels_
df_mm_sc
```

```python
df_mm_sc.groupby(['cluster'])['ash'].count()
```

##RobustScaler

```python
rb_sca = RobustScaler()
rb_sca.fit(df)
```

```python
tr_rb_sca = rb_sca.transform(df)
tr_rb_sca
```

```python
df_rb_sc= pd.DataFrame(tr_rb_sca, columns = df.columns)
df_rb_sc
```

```python
kmeans.fit(df_rb_sc)
print(kmeans.labels_)
```

```python
df_rb_sc['cluster'] = kmeans.labels_
df_rb_sc
```

```python
df_rb_sc.groupby(['cluster'])['ash'].count()
```
